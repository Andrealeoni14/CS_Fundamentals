### Libraries and Imports

```python
import pandas as pd
import numpy as np
import datetime
pd.set_option('display.max_columns', 50)

dogs = pd.read_csv(r"C:\Users\leonind\OneDrive - Gewiss S.p.A\Documenti\WORK\FCS_Project\adoptions\dogs.csv")
travels = pd.read_csv(r"C:\Users\leonind\OneDrive - Gewiss S.p.A\Documenti\WORK\FCS_Project\adoptions\dogTravel.csv")
states = pd.read_csv(r"C:\Users\leonind\OneDrive - Gewiss S.p.A\Documenti\WORK\FCS_Project\adoptions\NST-EST2021-POP.csv", names=['state', 'population'])
```

### 3. For each (primary) breed, determine the ratio between the number of dogs of Mixed Breed and those not of Mixed Breed. Hint: look at the secondary_breed

```python
# Groups the dogs by primary breed
primaryBreedGroup = dogs.groupby(['breed_primary'])

# Computes the percentage ratio of mexid breed per primary breed
mixedDogsRatioTable = pd.DataFrame(primaryBreedGroup['breed_mixed'].value_counts(normalize=True))
mixedDogsRatioTable.columns = ['ratio']
mixedDogsRatioTable
```

### 4. For each (primary) breed, determine the earliest and the latest posted timestamp

```python
# Finds the last and min 'posted' for every primary breed
maxPosted = pd.Series(primaryBreedGroup['posted'].last())
minPosted = pd.Series(primaryBreedGroup['posted'].min())

# Unites the two series in one dataframe
postedTimestampTable = pd.concat([minPosted, maxPosted], axis=1)
postedTimestampTable.columns = ['earliest posted', 'latest posted']
postedTimestampTable
```

### 5. For each state, compute the sex imbalance, that is the difference between male and female dogs. In which state this imbalance is largest?

```python
# Conts the number of female and male for every state
stateSexGroup = dogs.groupby(['contact_state'])
sexDogsState = pd.DataFrame(stateSexGroup['sex'].value_counts())
sexDogsState.columns = ['count']

# Pivots the table to have 'female' and 'male' as headers
sexDogsState = pd.pivot_table(sexDogsState, values='count', index=['contact_state'], columns=['sex'])
# Computs the difference between male and female into the 'Result' column
sexDogsState['Result'] = abs(sexDogsState['Male'] - sexDogsState['Female'])
# Shots where the difference is the highest
pd.DataFrame(sexDogsState.loc[sexDogsState['Result'].idxmax()]).columns
```

### 6. For each pair (age, size), determine the average duration of the stay and the average cost of stay.

```python
# Calculates the average stay duration and cost for every dog 'age' and 'size' combination
ageSizeGroup = dogs.groupby(['age', 'size'])
meanStay = ageSizeGroup['stay_duration'].mean()
meanCost = ageSizeGroup['stay_cost'].mean()

# Combines the results in one dataframe
pd.concat([meanStay, meanCost], axis=1)
```

### 7. Find the dogs involved in at least 3 travels. Also list the breed of those dogs.

```python
# Merges the dataframes 'dogs' and 'travels'
mergedDogsTravels = pd.merge(dogs, travels, on='id')
# Counts the travels for eache dog
countDogsTravels = pd.DataFrame(mergedDogsTravels['id'].value_counts())
# Keeps the merged dataframe of 'dogs' and 'travels' with only the rows of dogs with >= 3 travels
countDogsTravels = countDogsTravels.loc[countDogsTravels['id'] >= 3]
dogsWith3Travels = mergedDogsTravels.loc[mergedDogsTravels['id'].isin(countDogsTravels.index)]

# Extracts the 'id' and 'breed_primary' from the last result and then counts again the number of travels per dog
dogsWith3Travels = dogsWith3Travels[['id', 'breed_primary']]
dogsWith3Travels = pd.DataFrame(dogsWith3Travels.groupby(['id', 'breed_primary']).value_counts())
dogsWith3Travels.columns = ['number of travels']
dogsWith3Travels
```

### 8. Fix the travels table so that the correct state is computed from the manual and the found fields. If manual is not missing, then it overrides what is stored in found.

```python
# Changes the 'found' column value to the value of the column 'manual' if the column 'manual' is not null
travels['found'] = np.where(travels['manual'].isnull() == False, travels['manual'], travels['found'])
travels
```

### 9. For each state, compute the ratio between the number of travels and the population.

```python
mergedTravelsStates = pd.merge(states, travels, left_on='state', right_on='found', how = "left")

mergedTravelsStates['counting'] = np.where(mergedTravelsStates['found'].isnull() == False, 1, 0)

countTravelsStates = pd.DataFrame(mergedTravelsStates.groupby(['state', 'population'])['counting'].sum())
countTravelsStates.columns = ['number_of_travels'] 
countTravelsStates['ratio'] = countTravelsStates['number_of_travels']*2
countTravelsStates
```

### 10. For each dog, compute the number of days from the posted day to the day of last access.

```python
dogs["accessed"] = pd.to_datetime(dogs["accessed"], utc=True)

dogs['numberOfDays'] = (dogs["accessed"] - dogs['posted']).dt.days+1
dogs[['id', 'numberOfDays']]
```

### 11. Partition the dogs according to the number of weeks from the posted day to the day of last access.

```python
dogs['numberOfWeeks'] = (dogs['numberOfDays'] / 7).round()
weekGroup = dogs.groupby(['numberOfWeeks'])
pd.DataFrame(weekGroup['id'].value_counts())
```
