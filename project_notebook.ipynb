### 5. For each state, compute the sex imbalance, that is the difference between male and female dogs. In which state this imbalance is largest?

```python
# Conts the number of female and male for every state
stateSexGroup = dogs.groupby(['contact_state'])
sexDogsState = pd.DataFrame(stateSexGroup['sex'].value_counts())
sexDogsState.columns = ['count']

# Pivots the table to have 'female' and 'male' as headers
sexDogsState = pd.pivot_table(sexDogsState, values='count', index=['contact_state'], columns=['sex'])

# Computs the difference between male and female into the 'Result' column
sexDogsState['Result'] = abs(sexDogsState['Male'] - sexDogsState['Female'])

# Shots where the difference is the highest
pd.DataFrame(sexDogsState.loc[sexDogsState['Result'].idxmax()]).columns
```

### 8. Fix the travels table so that the correct state is computed from the manual and the found fields. If manual is not missing, then it overrides what is stored in found.

### 9. For each state, compute the ratio between the number of travels and the population.

```python
# Left joins the table states and travels
mergedTravelsStates = pd.merge(states, travels, left_on='state', right_on='found', how = "left")

# Counts the number of travels per state
mergedTravelsStates['counting'] = np.where(mergedTravelsStates['found'].isnull() == False, 1, 0)
countTravelsStates = pd.DataFrame(mergedTravelsStates.groupby(['state', 'population'], as_index = False)['counting'].sum())

# Removes the dots from the population number in order to convert the object to a numeric value
countTravelsStates['population'] = [x.replace('.', '') for x in countTravelsStates['population']]
countTravelsStates['population'] = countTravelsStates['population'].astype(int)

# Computes the ratio between 'population' and 'counting'
countTravelsStates['ratio'] = np.where(countTravelsStates['counting'] != 0, countTravelsStates['counting'] / countTravelsStates['population'], 0)
countTravelsStates
```

### 10. For each dog, compute the number of days from the posted day to the day of last access.

```python
# Transforms accessed to datetime
dogs["accessed"] = pd.to_datetime(dogs["accessed"], utc=True)

# Splits posted and then stores the detetime of the day into postedNew
dateSplitted = dogs["posted"].str.split("T", n = 1, expand = True)
dogs["postedNew"] = dateSplitted[0]
dogs["postedNew"] = dogs["postedNew"].astype(str)
dogs["postedNew"] = pd.to_datetime(dogs["postedNew"], errors ='coerce',  utc=True)

# Computs the number of days between posted and accessed
dogs['numberOfDays'] = (dogs["accessed"] - dogs['postedNew']).dt.days
dogs[['id', 'accessed', 'postedNew', 'numberOfDays']]
```

### 11. Partition the dogs according to the number of weeks from the posted day to the day of last access.

```python
# Creates a new column with the number of weeks
dogs['numberOfWeeks'] = (dogs['numberOfDays'] / 7).round()

# Groups by in order to create groups based on the number of weeks
weekGroup = dogs.groupby(['numberOfWeeks', 'postedNew', 'accessed'])
pd.DataFrame(weekGroup['id'].value_counts())
```

### 12. Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates.

```python
# Keeps all dogs with same breeds and sex
duplicatesDogs = dogs[['breed_primary', 'breed_secondary', 'breed_mixed', 'sex']].duplicated(keep=False)
duplicatesDogs = dogs[duplicatesDogs]
duplicatesDogs = duplicatesDogs[duplicatesDogs['description'].isnull() == False] # Excludes NaN values

# Method that returns true is the two strings share at leat 90% of the words
def similarWordsRario(s0, s1):
    s0 = s0.lower()
    s1 = s1.lower()
    s0List = s0.split(" ")
    s1List = s1.split(" ")
    similarCount = len(list(set(s0List)&set(s1List)))
    if len(s0.split()) > len(s1.split()):
        ratio = similarCount / len(s0.split())
    ratio = similarCount / len(s1.split())
    if ratio >= 0.9:
        return True
    return False

# Adds the descriptions with at least 90& similar words to an other to the 'rowsToKeep' variable
rowsToKeep = []
for indexX in duplicatesDogs.index:
    for indexY in duplicatesDogs.index:
        if indexX != indexY:
            if similarWordsRario(duplicatesDogs['description'][indexX], duplicatesDogs['description'][indexY]):
                if duplicatesDogs['description'][indexX] not in rowsToKeep:
                    rowsToKeep.append(duplicatesDogs['description'][indexX])
                if duplicatesDogs['description'][indexY] not in rowsToKeep:
                    rowsToKeep.append(duplicatesDogs['description'][indexY])

# Prints the dogs matching the request
duplicatesDogs[duplicatesDogs['description'].isin(rowsToKeep)]
```
